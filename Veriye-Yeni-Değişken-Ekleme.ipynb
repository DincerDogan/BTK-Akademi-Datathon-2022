## train test verilerinin alınması

```python
train = pd.read_csv("~/Desktop/DataThon_Yarisma/train.csv", parse_dates= ["tarih"])
test = pd.read_csv("~/Desktop/DataThon_Yarisma/testFeatures.csv", parse_dates= ["tarih"])

trainVeri = train.copy()
testVeri = test.copy()
```

```python
trainVeri = trainVeri.set_index("tarih")
trainVeri.head()
```

```python
testVeri = testVeri.set_index("tarih")
testVeri.head()
```

## Train ve test setine usd eklenmesi

```python
trainVeri = trainVeri.join(trainUSD)
```

```python
testVeri = testVeri.join(testUSD)
```

```python
trainVeri[4000:4005]
```

```python
trainVeri.drop(["Open", "High", "Low", "Change %"], axis=1, inplace=True)
```

```python
testVeri.drop(["Open", "High", "Low", "Change %"], axis=1, inplace=True)
```

### Usd ekleme kısmı

```python
trainVeri.info()
```

```python
NaN_price_list = trainVeri[trainVeri['Price'].isna()].index.unique()
NaN_price_list
```

#### Traine usd ekleme

```python
for weekend in NaN_price_list:
    trainVeri.loc[weekend, "Price"] = trainUSD.loc[weekend - timedelta(2), "Price"]
```

#### test verisine usd ekleme
